Q1: How many customers exist per city, and what is their average account balance?

```python
from tabulate import tabulate

# use a single SQL query that joins customers and accounts on customer_id
sql = """
SELECT c.city, c.name, a.account_balance
FROM customers_data AS c
JOIN accounts_data AS a ON c.customer_id = a.customer_id
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["city", "name", "avg_account_balance"]
print(tabulate(result, headers, tablefmt="grid"))
```

    Q2: Which account type (Savings, Current, Loan, etc.) holds the highest total balance?

```python
sql = """
SELECT c.account_type, SUM(a.account_balance) AS total_balance
FROM customers_data AS c
JOIN accounts_data AS a ON c.customer_id = a.customer_id
GROUP BY c.account_type
"""

cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["account_type", "total_balance"]
print(tabulate(result, headers, tablefmt="grid"))
```

Q3: Who are the top 10 customers by total account balance across all account types? in python

```python
sql_q3 = """
SELECT
    c.customer_id,
    CONCAT(c.name, ' ', c.name) AS customer_name,
    a.account_balance AS total_balance
FROM customers_data AS c
JOIN accounts_data AS a
    ON c.customer_id = a.customer_id
GROUP BY c.customer_id, customer_name
ORDER BY total_balance DESC
LIMIT 10;
"""
cursor_mysql.execute(sql_q3)
result = cursor_mysql.fetchall()
headers = ["customer_id", "customer_name", "total_balance"]
print(tabulate(result, headers, tablefmt="grid"))
```

Q4: Which customers opened accounts in 2023 with a balance above 1,00,000?


```python
sql = """
SELECT
    c.customer_id,
    c.name,
    c.join_date AS opening_date,
    a.account_balance
FROM
    customers_data AS c
JOIN
    accounts_data AS a
ON
    c.customer_id = a.customer_id
WHERE
    YEAR(c.join_date) = 2023
    AND a.account_balance > 100000;
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["customer_id", "customer_name", "opening_date", "account_balance"]
print(tabulate(result, headers, tablefmt="grid"))
```

Q5: What is the total transaction volume (sum of amounts) by transaction type?


```python
sql = """
SELECT
    t.txn_type,
    SUM(t.amount) AS total_transaction_volume
FROM
    transactions_data AS t
GROUP BY
    t.txn_type;
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["transaction_type", "total_transaction_volume"]
print(tabulate(result, headers, tablefmt="grid"))
```

    Q6: Which accounts have more than 3 failed transactions in a single month?


```python
sql = """
SELECT
    customer_id,
    DATE_FORMAT(txn_time, '%Y-%m') AS txn_month,
    COUNT(*) AS failed_count
FROM
    transactions_data
WHERE
    status = 'Failed'
GROUP BY
    customer_id,
    DATE_FORMAT(txn_time, '%Y-%m')
HAVING COUNT(*) > 3;
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["customer_id", "txn_month", "failed_count"]
print(tabulate(result, headers, tablefmt="grid"))
```

Q7: Which are the top 5 branches by total transaction volume in the last 6 months?




```python
sql = """
SELECT
    b.branch_id,
    COUNT(*) AS total_transactions
FROM
    transactions_data AS t
JOIN
    branches_data AS b
WHERE
    t.txn_time >= DATE_SUB(CURDATE(), INTERVAL 6 MONTH)
GROUP BY
    b.branch_id
ORDER BY
    total_transactions DESC
LIMIT 5;
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["branch_id", "total_transactions"]
print(tabulate(result, headers, tablefmt="grid"))
```

Q8: Which accounts have 5 or more high-value transactions above 2,00,000?


```python
sql = """
SELECT
    c.customer_id,
    c.name,
    COUNT(*) AS high_value_txn_count
FROM transactions_data AS t
JOIN customers_data AS c
    ON t.customer_id = c.customer_id
WHERE
    t.amount > 200000
GROUP BY
    c.customer_id, c.name
HAVING
    high_value_txn_count >= 0;
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["customer_id", "customer_name", "high_value_txn_count"]
print(tabulate(result, headers, tablefmt="grid"))
```

Q9: What is the average loan amount and interest rate by loan type (Personal, Auto, Home, etc.)?


```python
sql = """
SELECT
    loan_type,
    AVG(loan_amount) AS avg_loan_amount,
    AVG(interest_rate) AS avg_interest_rate
FROM
    loans_data
GROUP BY
    loan_type;
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["loan_type", "avg_loan_amount", "avg_interest_rate"]
print(tabulate(result, headers, tablefmt="grid"))
```

Q10: Which customers currently hold more than one active or approved loan?


```python
sql = """
SELECT
    c.customer_id,
    c.name AS customer_name,
    COUNT(l.Loan_ID) AS total_loans
FROM customers_data AS c
JOIN loans_data AS l
    ON c.customer_id = l.Customer_ID
WHERE l.Loan_Status IN ('Active', 'Approved')
GROUP BY c.customer_id, c.name
HAVING COUNT(l.Loan_ID) > 1;
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["customer_id", "customer_name", "total_loans"]
print(tabulate(result, headers, tablefmt="grid"))
```

Q11: Who are the top 5 customers with the highest outstanding (non-closed) loan amounts?


```python
sql = """
SELECT
    c.customer_id,
    c.name AS customer_name,
    SUM(l.Loan_Amount) AS total_outstanding_amount
FROM customers_data c
JOIN loans_data l
    ON c.customer_id = l.Customer_ID
WHERE l.Loan_Status <> 'Closed'
GROUP BY c.customer_id, c.name
ORDER BY total_outstanding_amount DESC
LIMIT 5;
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["customer_id", "customer_name", "total_outstanding_amount"]
print(tabulate(result, headers, tablefmt="grid"))
```

Q13: What is the branch performance summary showing total customers, total loans, and transaction volume?


```python
sql = """
SELECT
    b.branch_id,
    b.branch_name,

    -- Total customers in this branch
    COUNT(DISTINCT c.customer_id) AS total_customers,

    -- Total loans issued by this branch
    COUNT(DISTINCT l.loan_id) AS total_loans,

    -- Total transaction volume
    COALESCE(SUM(t.amount), 0) AS total_transaction_volume

FROM branches_data AS b
LEFT JOIN customers_data AS c
    ON b.branch_id

LEFT JOIN accounts_data AS a
    ON c.customer_id = a.customer_id

LEFT JOIN transactions_data AS t
    ON a.customer_id = t.customer_id

LEFT JOIN loans_data AS l
    ON c.customer_id = l.customer_id

GROUP BY
    b.branch_id, b.branch_name
ORDER BY
    b.branch_id;
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["branch_id", "branch_name", "total_customers", "total_loans", "total_transaction_volume"]
print(tabulate(result, headers, tablefmt="grid"))
```

Q14: Which issue categories have the longest average resolution time?


```python
sql = """
SELECT
    Issue_Category,
    AVG(TIMESTAMPDIFF(DAY, Date_Opened, Date_Closed)) AS avg_resolution_days
FROM support_tickets_data
WHERE Date_Closed IS NOT NULL
GROUP BY Issue_Category
ORDER BY avg_resolution_days DESC;
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["issue_category", "avg_resolution_days"]
print(tabulate(result, headers, tablefmt="grid"))
```

Q15: Which support agents have resolved the most critical tickets with high customer ratings (4)?


```python
sql = """
SELECT
    Support_Agent AS agent_name,
    COUNT(Ticket_ID) AS resolved_critical_high_rated_tickets
FROM
    support_tickets_data
WHERE
    Priority = 'Critical'
    AND Status = 'Resolved'
    AND Customer_Rating >= 4
    AND Support_Agent IS NOT NULL
GROUP BY
    Support_Agent
ORDER BY
    resolved_critical_high_rated_tickets DESC;
"""
cursor_mysql.execute(sql)
result = cursor_mysql.fetchall()
headers = ["agent_name", "resolved_critical_high_rated_tickets"]
print(tabulate(result, headers, tablefmt="grid"))
```
